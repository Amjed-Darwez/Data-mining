## 8 Observer la collone `Author`
### Completer la fonction `clean_author_names`qui permet de corriger les valeurs de cet colonne

````python
# compléter la fonction suivante ( à la place de pass)


def clean_author_names(author):
    author = str(author)

    if author == 'nan':
        return 'NaN'

    author = author.split(',')

    if len(author) == 1:
        name = filter(lambda x: x.isalpha(), author[0])
        return reduce(lambda x, y: x + y, name)

    last_name, first_name = author[0], author[1]

    first_name = first_name[:first_name.find('-')] if '-' in first_name else first_name

    if first_name.endswith(('.', '.|')):
        parts = first_name.split('.')
        if len(parts) > 1:
            first_name = parts[0].strip()
        else:
            first_name = first_name.strip()


    first_name = first_name.strip() # Remove leading/trailing whitespaces
    last_name = last_name.strip().capitalize() # Capitalize last name

    return f'{first_name} {last_name}'


df['Author'] = df['Author'].apply(clean_author_names)
df.head() # To see the output, run the code.
````

# Résultat attendu

````python
## complétrer la fonction suivante ( à la place de pass)


def clean_title(title):

    if title == 'nan':
        return 'NaN'

    if title[0] == '[':
        title = title[1:title.find(']')].strip()

    if 'by' in title:
        title = title[:title.find('by')].strip()
    elif 'By' in title:
        title = title[:title.find('By')].strip()

    if '[' in title:
        title = title[:title.find('[')].strip()

    title = list(map(str.capitalize, title.split()))
    return ' '.join(title)

df['Title'] = df['Title'].apply(clean_title)
df.head()
````

## 2 Créer un pandas data frame à partir d'un fichier txt
On veut créer un DataFrame nommé `towns_df` à partir du fichier 'university_towns.txt'

Nous voyons que nous avons des noms d'états périodiques suivis des villes universitaires de cet état: StateA TownA1 TownA2 StateB TownB1 TownB2 .... Si nous regardons la façon dont les noms d'états sont écrits dans le fichier, nous verrons qu'ils ont tous la sous-chaîne «[edit]» en eux.

Nous pouvons tirer parti de ce modèle en créant une liste de tuples (état, ville) et en enveloppant cette liste dans un DataFrame:


````python
university_towns = []
state = None  # Initialize state to None
with open('/content/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            # Remember this `state` until the next is found
            state = line[:-7]  # Remove '[edit]\n' from state name
        else:
            # Otherwise, we have a town; keep `state`
            university_towns.append((state, line.strip()))  # Append (state, town) tuple

university_towns[:5] # To see the output, run the code.
````

````python
towns_df = pd.DataFrame(university_towns,
                        columns=['State', 'RegionName'])
towns_df.head()
````

## 3 Compléter la fonction `clean_up` pour nettoyer les colonnes de `towns_df`

````python
## complétrer la fonction suivante ( à la place de pass)
def clean_up(item):
    if '(' in item:

        return item[:item.find('(') - 1]

    if '[' in item:
        return item[:item.find('[')]


    return item


towns_df = towns_df.applymap(clean_up)
towns_df
````

````python
Q1 = np.quantile(olympics_df['Total Medals'], 0.25)
Q2 = np.quantile(olympics_df['Total Medals'], 0.50)  # Also the median
Q3 = np.quantile(olympics_df['Total Medals'], 0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}, Q2: {Q2}, Q3: {Q3}, IQR: {IQR}")
````

````python
# Calculate upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Identify outliers
outliers = olympics_df[(olympics_df['Total Medals'] > upper_bound) | (olympics_df['Total Medals'] < lower_bound)]

# Number of outliers
num_outliers = len(outliers)
print(f"Total number of outliers: {num_outliers}")
````
